```python
import lark
```

```python
grammar = """
start: bars

bars: header? lines (header lines)*
header: SEPARATOR CNAME?
lines: line (NEWLINE line)*
line: group*
group: note | "(" group+ ")" time

note: ABSOLUTE? octave? CNAME augmentation? dot_duration?
octave: (OCTAVE_DOWN)+ | (OCTAVE_UP)+
augmentation: SIGN INT
dot_duration: (".")+

%import common.INT
time: INT | INT "/" INT

OCTAVE_DOWN: "<"
OCTAVE_UP: ">"
SEPARATOR: "-" "-" "-"+
ABSOLUTE: "@"
SIGN: "+" | "-"

%import common.NEWLINE
%import common.CNAME

%import common.WS_INLINE
%ignore WS_INLINE

%import common.SH_COMMENT
%ignore SH_COMMENT
"""

parser = lark.Lark(grammar)

print(parser.parse("""
# stuff
--- first
do... >do @<<do
(what)2 (ever this-2 is)4/3
----
first+2
""").pretty())
```
